```python
import random  
import numpy as np
import time as tm

start = tm.clock()


userclass=[1,2,3]
        
B=np.zeros(3,dtype=np.float)
        
D=np.zeros(3,dtype=np.float)
       
L=np.zeros(3,dtype=np.float)
        
It=np.zeros(3,dtype=np.float)
        
Ie=np.zeros(3,dtype=np.float)
        
for i in range(3):
   if userclass[i]==1:
      B[i]=random.uniform(10,20)
      D[i]=1000
      L[i]=0.3
      It[i]=0.7
      Ie[i]=0.3
            
   elif userclass[i]==2:
      B[i]=random.uniform(4,8)
      D[i]=400
      L[i]=0.7
      It[i]=0.5
      Ie[i]=0.5              
   else:
      B[i]=random.uniform(1,2)
      D[i]=100
      L[i]=1
      It[i]=0.3
      Ie[i]=0.7
        
A=np.zeros(3,dtype=np.float)
        
T=np.zeros(3,dtype=np.float)
E=np.zeros(3,dtype=np.float)
COST=np.zeros(3,dtype=np.float)
Rt=np.array([1,1,1])
Rb=np.array([1,1,1])
Pt=np.array([0.1,0.1,0.1])
Pb=np.array([0.01,0.01,0.01])
f=2.5
z=10**(-26)*f**2
TC=np.zeros(100,dtype=np.float)
CAP=np.zeros(100,dtype=np.float)
Noff=np.zeros(100,dtype=np.float)
FULL=15
F=5
all=[]     
for j in range(100):
    A=([random.randint(0,1),random.randint(0,1),random.randint(0,1)])
    total=0
    off=0
    print (A)
    for i in range(3):
        if A[i]==0:
           T[i]=D[i]/f
           E[i]=z*D[i]
        else :
           T[i]=B[i]/Rt[i]+D[i]/F+B[i]/(2*Rb[i])
           E[i]=Pt[i]*B[i]/Rt[i]+Pb[i]*B[i]/(2*Rb[i])
           off+=1
        COST[i]=It[i]*T[i]+Ie[i]*E[i]
        total+=COST[i]
        
    Noff[j]=off
    TC[j]=total  
    CAP[j]=FULL-Noff[j]*F
    if CAP[j]<0:
        continue 
    all=[A,Noff,TC,CAP] 
    end = tm.clock()

print (all)

print(end-start)
```
